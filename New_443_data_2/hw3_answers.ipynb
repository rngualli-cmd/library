*Answer*: There are 30 samples.  To compute a quantile we first order the samples.  
We represent the n unordered samples as 

$$x_1, x_2, \dots, x_n$$

We sort them resulting in

$$x_{(1)} \leq x_{(2)} \leq \dots \leq x_{(n)}$$

Thus,

$$x_{{1}}$$ is the minimum, which is also called the 1st order statistic.

$$x_{(n)}$$ is the maximum, which is also called the nth order statistic.

$$x_{(k)}$$ is the kth smallest and is thus called the kth order statistic.

When we are looking for the median, and we have an odd number of samples, we know
that median has an equal number of samples above as below and the median is the middle sample value.   When we have an even number, the median appears as the
average of the two middle samples.


$$
\\tilde{x} = \\text{median}(x) =
\\begin{cases}
x_{\\left(\\frac{n+1}{2}\\right)}, & \\text{if } n \\text{ is odd} \\\\
\\\\
\\dfrac{x_{\\left(\\frac{n}{2}\\right)} + x_{\\left(\\frac{n}{2}+1\\right)}}{2}, & \\text{if } n \\text{ is even}
\\end{cases}
$$

This means the median is the average between the 15th and 16th order statistic.  The 15th and 16th fall within the bucket spanning the range [0.7, 0.8]. 

Statisticians sometimes assume the distrbution is 
uniform within the bucekt yield a median that is approximately 0.75. 

Since we do not know the distribution of samples within a bucket, the tightest justifiable
bounds on the median spans the bucket so 

$$\boxed{\tilde{x} \in [0.7, 0.8]}$$.

**3.2** A random variable obeys the following continuous probability density function. What is its mean?

$$f(x) = 0.1 e^{-0.1x}$$

A: When dealing with sample mean, the mean is given by 

  $$\overline{x} = \frac{1}{n}\sum_{i=1}^n x_i$$

  For a distribution involving a continuous distribution, the distribution mean is given by

  $$\mu = \int_{x=-\infty}^{\infty} x f(x) dx $$

  For Problem 3.2, we are given a continuous probability density function and asked to compute the distribution mean. In order for this to be a legitimate probability density function, its integral from negative infinity to infinity must equal 1. If we compute
  
  $$\int_{x=-\infty}^\infty 0.1 e^{-0.1x} dx$$

  we would find that not only is this not equal to 1, but it diverges in the range of *x< 0*. *f(x)* is only a legitimate probability density function for *x > 0* and this would be consistent with *x* being an exponential random variable.
  Thus, when computing the mean, we only consider the range *x>0*.

  $$\mu = \int_{x=0}^{\infty} x 0.1 e^{-0.1x} dx = 10$$

  $$\boxed{\mu=10}$$

**3.3** A random variable obeys the following continuous probability density function. What is its distribution variance?

$$\frac{2(x-a)}{(b-a)(c-a)} \text{ if } a \leq x < c$$
$$\frac{2(b-x)}{(b-a)(b-c)} \text{ if } c \leq x \leq b$$
$$0 \text{ otherwise}$$

A: Let's plug in some arbitrary values for *a*, *b*, and *c* and plot it.  I arbitrarily chose *a=0, b=3, c=1*.

```python
import matplotlib.pyplot as plt
import numpy as np

a = 0
b = 3
c = 1

def f(x):
    if x < a:
        return 0
    elif a <= x < c:
        return 2 * (x-a)/ ((b-a)*(c-a))
    elif c < x <= b:
        return 2 * (b-x) / ((b-a)*(b-c))
    else:
        return 0
    
# generate points equally spaced along the real line throughout the range [a, b].
x_values = np.linspace(a, b, 1000) 

y_values = [f(x) for x in x_values]

plt.plot(x_values, y_values)
plt.show()
```

This pdf is a triangle.  If this is a legitimate probability density function then (1) its area should be 1 and (2) all values should be non-negative.  The triangle lies above the x-axis so all values are non-negative, so (2) is satisfied.  Let's verify (1).

The area of a triangle is

$$\text{Area } = \frac{1}{2} (b-a) h$$

where *h* is the height.  To compute the height we plug $c$ into either equation for the edges of the triangle as specified in the definition of $f(x)$.

$$h = \frac{2(c-a)}{(b-a)(c-a)} = \frac{2}{b-a}$$

The area is thus

$$\text{Area } = \frac{1}{2} (b-a) \frac{2}{b-a} = 1$$

Both necessary conditions for a probability density function are satsified.  This is a legitimate PDF.

Equation 1:

Given that we know this is a triangle, we can approach this problem at least two different ways.
One is to use calculus, which takes a bit of doing, or we break the triangle in two and
treat it as two right triangles.

If we balance a right triangle on one edge, other than its hypotenuse, the mean is 1/3 of the distance along the edge from the right angle.  We can thus compute the mean of each of the right triangles and then take the weighted mean of the two where we weight the means in proportion to the relative areas of the triangles.

Note if you don't want to believe this analysis, I will do it using calculus as well in a moment.

Let *L* denote the right-triangle on the left.  It's bottom edge spans from *x=a* to *x=c*.  

Let *R* denote the right-triangle on the right.  It's bottom edge spans from *x=c* to *x=b*.

The mean of the left triangle is two thirds of the way from *a* to *c*, i.e., 

$$\text{mean}(L) = \frac{1}{3} a + \frac{2}{3} c$$

$$\text{mean}(R) = \frac{2}{3} c + \frac{1}{3} b$$

We can then compute the mean of the means weighted by their relative areas

$$\mu = \frac{\text{area}(L) \cdot \text{mean}(L) + \text{area}(R) \cdot \text{mean}(R)}{\text{area of overall triangle}}$$

Since this is a PDF, the area of the overall triangle is known to be 1, so the equation above simplifes to
 
$$\mu = \text{area}(L) \cdot \text{mean}(L) + \text{area}(R) \cdot \text{mean}(R)$$

$$\text{area}(L) = \frac{1}{2}(c-a)\cdot h$$

$$\text{area}(R) = \frac{1}{2}(b-c)\cdot h$$

We already calculated *h* which can substitute into the two equations above to get 

$$\text{area}(L) = \frac{1}{2}(c-a)\cdot \frac{2}{b-a} = \frac{c-a}{b-a}$$

$$\text{area}(R) = \frac{1}{2}(b-c)\cdot \frac{2}{b-a} = \frac{b-c}{b-a}$$

The mean is thus given by 

$$\mu = \frac{c-a}{b-a} \big(\frac{1}{3} a + \frac{2}{3} c\big) + \frac{b-c}{b-a}\big(\frac{2}{3} c + \frac{1}{3}b\big)$$

which simplifies to

$$\mu = \frac{a+b+c}{3}$$

Variance is defined as

$$
\text{Var}[X] = E[(X-E[X])^2] 
$$

where *E[X]* is the expected value for X which is μ.

This can be rearranged to

$$
\text{Var}[X] = E[X^2] - \mu^2 
$$

To compute the distribution variance, we do use Equation 3 below

$$
\text{Var}[X] = \int_{x=-\infty}^{\infty} x^2 f(x) dx - \mu^2
$$

Equation 4 significantly reduces the amount of work we have to do compared 
to Equation 3.  I suggest memorizing Equation 4.  It is a much simpler 
version of Equation 3.  Substituting in our definitions for *f(x)* yields

$$
\text{Var}[X] = \int_{x=a}^{c} x^2 \frac{2(x-a)}{(b-a)(c-a)} dx+ \int_{x=c}^{b} x^2 \frac{2(b-x)}{(b-a)(b-c)} dx - \mu^2
$$

*a*, *b*, and *c* are all constants, so we can rearrange the equation above to Equation 5 below

$$
\text{Var}[X] = \frac{2}{(b-a)(c-a)} \int_{x=a}^{c} x^2 (x-a) dx + \frac{2}{(b-a)(b-c)} \int_{x=c}^{b} x^2 (b-x) dx - \mu^2

Which with a significant amount of algebra yields

For any random variable the distribution mean can be computed using an integral
as follows:

$$\mu = \int_{x=-\infty}^{+\infty} x f(x) dx$$

For a uniform random variable U[a,b], the probability is constant for all values
within the range [a,b], and for every probability density function, the integral 
across the real-line must equal 1.  If the probability density is equal for all
values in the range [a,b] then for the integral across the range [a,b] to be 1 then the

$$f(x) = \frac{1}{40-20} \text{~~~~~~~if } 20 \leq x \leq 40$$

$$\mu = \int_{x=-\infty}^{+\infty} x f(x) dx = \int_{x=20}^{40} x \cdot \frac{1}{40-20} dx$$

$$\mu = \frac{1}{20} \left. \frac{x^2}{2} \right|_{20}^{40}$$

$$\boxed{\mu=30}$$

5.2 A manufacturing process produces steel rods that are supposed to be 100 cm in length. What is the probability that a randomly selected steel rod from this process will be either shorter than 96 cm or longer than 104 cm? Use the error function (erf) to compute this probability. Show in 
steps how you came to this answer.  Practice using LaTeX.

A: We want to find the 
$$P[X<96 \text{ or } X > 104]$$

The Z-scores for samples landing on 96 or 104 are given by Equations 1 and 2 below

$$Z_{96}=\frac{96-100}{2} = -2$$

$$Z_{104} = \frac{104-100}{2} = 2$$

The Z-score is a measure of how many standard deviations a sample is away from 
mean.  The probability of a sample falling within Z standard deviations is given by
Equation 3 below

$$P[-z \leq Z \leq z] = \Phi(z) - \Phi(-z)$$

Equation 4 below is the relationship between the error function (erf) and Phi (Φ).

$$\Phi(Z) = \frac{1}{2}\big[ 1 + \text{erf}\big(\frac{Z}{\sqrt{2}})\big]$$ 

If we substitute the erf in Equation 4 into Equation 3, we get

$$P[-z \leq Z \leq z] = \frac{1}{2} \big[1 + \text{erf}(\frac{z}{\sqrt{2}})\big] - \frac{1}{2} \big[1 + \text{erf}(\frac{-z}{\sqrt{2}})\big]$$

$$P[-z \leq Z \leq z] = \cancel{\frac{1}{2}} + \frac{1}{2} \text{erf}(\frac{z}{\sqrt{2}}) - \cancel{\frac{1}{2}} -\frac{1}{2}\text{erf}(\frac{-z}{\sqrt{2}})$$

$$P[-z \leq Z \leq z] =  \frac{1}{2} \text{erf}(\frac{z}{\sqrt{2}}) - \frac{1}{2}\text{erf}(\frac{-z}{\sqrt{2}})$$

Because the error function is odd,

$$\text{erf}(x) = - \text{erf}(-x)$$

and thus we get Equation 5

$$P[-z \leq Z \leq z] =  \text{erf}(\frac{z}{\sqrt{2}})$$

Equation 5 shows how closely the error function is to the probabiilty that a Gaussian
random variable stays within *z* standard deviations of the mean.  Equation 5 is 
the same as saying

$$P[Z \text{ is within } z \text{ standard deviations of the mean}] = \text{erf}(\frac{z}{\sqrt{2}})$$

For this problem, we want the probability that the lengths of a beam go outside 2
standard deviations from the mean, i.e.,

$$P[Z < -z \text{ or } Z > z] = 1 - \text{erf}(\frac{2}{\sqrt{2}}) \approx 0.0455$$

There is about a 4.55% chance that a value will fall outside the range defined by 2 standard deviations.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

def plot_log_normal_with_sampling_distribution(m, n, mu, sigma, bins=20):
    # Generate m sample means, each computed from n samples drawn from a lognormal distribution
    sample_means = np.array(
        [np.mean(np.random.lognormal(mu, sigma, n)) for _ in range(m)])

    # Plot histogram of sample means
    plt.figure(figsize=(12, 8))
    plt.hist(sample_means, bins=bins, density=True, color='skyblue', edgecolor='black', alpha=0.7)

    # Generate x values for plotting the PDF
    x = np.linspace(0.001, 5, 500)  # Define range (adjust as needed)
    pdf = st.lognorm.pdf(x, s=sigma, scale=np.exp(mu))  # Compute PDF

    # Plot the Lognormal PDF
    plt.plot(x, pdf, 'r-', linewidth=2, label='Lognormal PDF')

    # Overlay a normal approximation for the sampling distribution 
    # using the Central Limit Theorem.  It isn't required
    # for the problem that we include the 
    # Compute theoretical mean and standard deviation of the lognormal distribution
    theoretical_mean = np.exp(mu + (sigma**2) / 2)
    theoretical_var = (np.exp(sigma**2) - 1) * np.exp(2 * mu + sigma**2)
    theoretical_std = np.sqrt(theoretical_var)

    # Compute the theoretical mean and standard error for the sampling distribution
    mean_of_means = theoretical_mean  # Expected mean of sample means
    std_of_means = theoretical_std / np.sqrt(n)  # Standard error (SE)

    # Generate normal approximation based on the central limit theorem.
    clt_pdf = st.norm.pdf(x, loc=mean_of_means, scale=std_of_means)
    plt.plot(x, clt_pdf, 'g-', linewidth=2, label='Theoretical Normal (CLT)')

    # Add a vertical line at the theoretical mean of the lognormal distribution
    plt.axvline(x=theoretical_mean, color='green', linestyle='dashed', linewidth=2, label="Lognormal Mean")

    plt.title('Lognormal Distribution and Sampling Distribution of the Sample Means')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

# Generate a lognormal distribution
# Set parameters for the lognormal distribution: mean, sigma, and number of samples
mu, sigma = 0, 0.5 # mean and standard deviation

# Sample size per sample mean
n = 5  # samples per sample mean
m = 100  # number of sample means

plot_log_normal_with_sampling_distribution(m, n, mu, sigma)
```
